```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import confusion_matrix
import joblib
```

```python
# Load the dataset
df = pd.read_csv(r'C:\Users\4dacc\Desktop\PractStan\Datasets\cs-training.csv')
```

```python
# Take a look at the first few rows
df.head()
```

```python
# Check for missing values and data types
df.info()
```

```python
# Set the visual style
sns.set(style="whitegrid")
```

```python

# Check the balance of the target variable
target_counts = df['SeriousDlqin2yrs'].value_counts()
target_counts
```

```python
# Print the numbers
print("Counts of Default (1) vs No Default (0):")
print(target_counts)
print("\nPercentage of Defaults:")
print(target_counts / len(df))
```

```python
# Visualize it
plt.figure(figsize=(6, 4))
sns.countplot(x='SeriousDlqin2yrs', data=df, palette='rocket')
plt.title('Distribution of Target Variable (Default vs No Default)')
plt.xlabel('Defaulted? (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# 1. Handle Missing Values
# We use the median for MonthlyIncome because it's robust against outliers (rich people skew the average)
imputer = SimpleImputer(strategy='median')

# Select columns to impute (you can just impute all for simplicity here, or specific ones)
cols_to_impute = ['MonthlyIncome', 'NumberOfDependents']
df[cols_to_impute] = imputer.fit_transform(df[cols_to_impute])

# Double-check we have no more nulls
print("Missing values after imputation:")
print(df.isnull().sum())
```

```python
# 2. Prepare for SMOTE
# Split features (X) and target (y)
# We drop 'Unnamed: 0' because it's just an index, not useful for prediction
X = df.drop(['SeriousDlqin2yrs', 'Unnamed: 0'], axis=1)
y = df['SeriousDlqin2yrs']

print("\nClass distribution BEFORE balancing:")
print(y.value_counts())
```

```python
# 3. Apply SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print("\nClass distribution AFTER balancing:")
print(y_res.value_counts())
```

```python
#Visualise
plt.figure(figsize=(6, 4))
sns.countplot(x=y_res, palette='rocket') # Try 'mako' or 'crest' here too!
plt.title('Distribution of Target Variable (AFTER SMOTE)')
plt.xlabel('Defaulted? (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# 1. Split the Data
# We use 20% of the data for testing, and 80% for training
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
# 2. Initialize the Model
# n_estimators=100 means it builds 100 decision trees
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# 3. Train the Model
print("Training the Random Forest model...")
rf_model.fit(X_train, y_train)
print("Training complete!")
```

```python
# 4. Make Predictions
y_pred = rf_model.predict(X_test)
```

```python
# 5. Evaluate the Model
print("\n--- Model Performance ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# 1. Generate the confusion matrix again (just to be safe)
cm = confusion_matrix(y_test, y_pred)

# 2. Create the heatmap
plt.figure(figsize=(8, 6))

# Define labels for the axes
labels = ['No Default (0)', 'Default (1)']

# Create the heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='rocket', 
            xticklabels=labels, yticklabels=labels)

plt.title('Confusion Matrix Heatmap', fontsize=16)
plt.ylabel('Actual Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
# 1. Define the grid of parameters to test
# We are telling Python: "Try depths between 10 and 50, try trees between 100 and 500..."
param_dist = {
    'n_estimators': randint(100, 500),
    'max_depth': randint(10, 50),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5)
}

# 2. Initialize the Randomized Search
# n_iter=20 means it will try 20 different random combinations.
# cv=3 means it uses 3-fold Cross-Validation (more robust testing).
rf_random = RandomizedSearchCV(estimator=rf_model, 
                               param_distributions=param_dist, 
                               n_iter=5, 
                               cv=3, 
                               verbose=1, 
                               random_state=42, 
                               n_jobs=1) # n_jobs=-1 uses all your CPU cores (faster)
```

```python
# This saves your trained model to a file named 'credit_model.pkl'
# It will appear in the same folder as your notebook
joblib.dump(rf_model, 'credit_model.pkl')

print("Success! Your model brain has been saved to the folder.")
```

```python
## 3. Fit the model
print("Starting the search for the best settings... (This might take a few mins)")
rf_random.fit(X_train, y_train)
```

```python
# 4. Check the best parameters
print("\nBest Parameters found:")
print(rf_random.best_params_)
```

```python
# 5. Evaluate
best_model = rf_random.best_estimator_
y_pred_best = best_model.predict(X_test)

print("\n--- New Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best):.4f}")
print(classification_report(y_test, y_pred_best))
```

```python
# Save the BEST model (the one we just tuned with the Safe Mode)
joblib.dump(best_model, 'credit_model.pkl')

print("The optimized model brain is saved!")
```
